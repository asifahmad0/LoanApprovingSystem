```python
def convert_to_model_input(data):

    employment_map = {'Salaried': 1, 'Self-employed': 0}
    married_map = {'Married': 1, 'Single': 0}
    loan_map = {'Personal': 1, 'Home': 0}
    area_map = {'Urban': 1, 'Rural': 0}
    edu_map = {'Graduate': 1, 'Not Graduate': 0}
    gender_map = {'Male': 1, 'Female': 0}
    sector_map = {'Private': 1, 'Government': 0}

    model_input = [
        data['Income'],
        data['Loan_Amount'],
        employment_map[data['Employment_Status']],
        data['Age'],
        married_map[data['Marital_Status']],
        data['Dependents'],
        data['Credit_Score'],
        data['Experience'],
        data['EMI_Ratio'],
        data['Savings'],
        data['Assets'],
        data['Liabilities'],
        data['Tenure'],
        loan_map[data['Loan_Purpose']],
        area_map[data['Property_Area']],
        edu_map[data['Education_Level']],
        gender_map[data['Gender']],
        sector_map[data['Employer_Category']]
    ]

    return model_input

```

```python
import pickle
import pandas as pd

# load model + encoder
model = pickle.load(open("model.pkl", "rb"))
ohe = pickle.load(open("encoder.pkl", "rb"))   # 👈 ye missing tha

def predict(input_dict):

    df = pd.DataFrame([input_dict])

    # categorical columns
    cat_cols = ['Gender','Employer_Category','Marital_Status',
                'Loan_Purpose','Property_Area','Employment_Status']

    # encode
    encoded = ohe.transform(df[cat_cols])
    encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())

    # numeric columns
    num_df = df.drop(columns=cat_cols)

    # final input
    final_df = pd.concat([num_df, encoded_df], axis=1)

    # prediction
    result = model.predict(final_df)

    return result
```

```python
input_data = {
    'Income': 17795.0,
    'Loan_Amount': 1387.0,
    'Employment_Status': 'Salaried',
    'Age': 51.0,
    'Marital_Status': 'Married',
    'Dependents': 0.0,
    'Credit_Score': 637.0,
    'Experience': 4.0,
    'EMI_Ratio': 0.53,
    'Savings': 19403.0,
    'Assets': 45638.0,
    'Liabilities': 16619.0,
    'Tenure': 84.0,
    'Loan_Purpose': 'Personal',
    'Property_Area': 'Urban',
    'Education_Level': 1,
    'Gender': 'Female',
    'Employer_Category': 'Private'
}


print(predict(input_data))
```
