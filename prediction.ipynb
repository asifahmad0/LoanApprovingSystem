```python
import pickle
import pandas as pd

# load model + encoder
model = pickle.load(open("nb_model.pkl", "rb"))
ohe = pickle.load(open("encoder.pkl", "rb"))   # 👈 ye missing tha

def predict(input_dict):

    df = pd.DataFrame([input_dict])

    # categorical columns
    cat_cols = ['Gender','Employer_Category','Marital_Status',
                'Loan_Purpose','Property_Area','Employment_Status']

    # encode
    encoded = ohe.transform(df[cat_cols])
    encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())

    # numeric columns
    num_df = df.drop(columns=cat_cols)

    # final input
    final_df = pd.concat([num_df, encoded_df], axis=1)

    # prediction
    result = model.predict(final_df)

    return result
```
